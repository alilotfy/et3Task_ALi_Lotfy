# Import dataset

```python
import pandas as pd

df = pd.read_csv('/content/drinkMenu.csv')
print (df)
```

```python
df.info()
```

# Remove Duplicate

```python
df_1=df.drop_duplicates()
df_1.info()
```

# Fill Null Values

```python
df_2=df_1.fillna(0)
df_2.info()
```

#Highest Calories : White Chocolate Mocha (Without Whipped Cream)




```python
df_2.sort_values('Calories', ascending=False, inplace=True)

df_3 = df_2.head(25)
```

```python
import matplotlib.pyplot as plt

labels = df_3["Beverage"]
counts = df_3["Calories"]


plt.subplot(111)
plt.plot(labels, counts)
plt.xticks(labels, labels, rotation='vertical')
plt.margins(0.2)
plt.title("Highest Calories")

plt.show()
```

#Highest Sugars: Java Chip (Without Whipped Cream)


```python
df_2.sort_values(' Sugars (g)', ascending=False, inplace=True)

df_4 = df_2.head(25)
labels = df_4["Beverage"]
counts = df_4[" Sugars (g)"]


plt.subplot(111)
plt.plot(labels, counts)
plt.xticks(labels, labels, rotation='vertical')
plt.margins(0.2)
plt.title("Highest sugars")

plt.show()
```
